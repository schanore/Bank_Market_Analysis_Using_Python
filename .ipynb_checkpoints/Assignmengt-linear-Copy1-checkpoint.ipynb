# Bank Marketing Data set

## Abstract

The data set is related with direct marketing campaigns of a Portuguese banking institution.Here we are trying to predict the balance and deposit variable which are the target variables using linear regression and logistic regression models. 
To convert the categorical data to integer values we are using the one hot encoding.Stepwise regression is used to build our regression model from a set of candidate predictor variables by entering and removing predictors â€” in a stepwise manner.We are also checking the multicollinearity to remove any biasing from the dataset.Cross Validation is mainly used to estimate how accurately a predictive model will perform in practice.Using L2 Regularization method which is also called as ridge regularization.To deal with the overfitting issues in the model this is being used.

We can see that the linear model 3 is showing the best results for predicting balance variable and logistic model 2 is 
giving the best results for predicting deposit.

For Linear regression model the best model is linear Regression model 3 with the R squared value as 0.201 and for logistic regression model the best model is linear regression model 2 with the accuracy of 0.76.

### Multicollinearity and Step-wise regression:

We need to check the multicollinearity so as to remove high correlation between independent variables. 
If the degree of correlation between variables is high enough, it can cause problems when you fit the model and interpret the results

To make the predictions from the model more accurate we will be using Stepwise regression.
In every step we will be removing one variable that is not affecting the target variable to a major extend.

```python
#performing stepwise collinearity on Linear model 3:
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["features"] = x.columns
vif.round(1)
```

*Is there any multi-colinearity in the model?*<br>
Yes there is some multicolinearity present in the model which will be removed by Step-wise regression.


*In the multiple regression models are predictor variables independent of all the other predictor variables?*<br>
In the prior model the independent variables were also dependent on other independent variables

*In multiple regression models rank the most significant predictor variables and exclude insignificant ones from the model*<br>
This is done using the below stepwise regression.


*Cross-validate the models. How well did they do?*<br>
The cross-validation for sure shows that better results compared to not validated ones but still the result is not that great.

```python
def rmsle(predicted, real):
    sum=0.0
    for x in range(len(predicted)):
        p = np.log(predicted[x]+1)
        r = np.log(real[x]+1)
        sum = sum + (p - r)**2
    return (sum/len(predicted))**0.5
```
